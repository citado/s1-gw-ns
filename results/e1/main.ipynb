```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## 1 pps

```python
r = pd.read_csv('./s1.csv', header=None)

r.head()
```

```python
avg = np.average(r, axis=1)
```

```python
s1avg = np.average(avg)
s1err = np.std(avg)
```

## 10 pps

```python
r = pd.read_csv('./ms100.csv', header=None)

r.head()
```

```python
avg = np.average(r, axis=1)
```

```python
ms100avg = np.average(avg)
ms100err = np.std(avg)
```

## 100 pps

```python
r = pd.read_csv('./ms10.csv', header=None)

r.head()
```

```python
avg = np.average(r > 0, axis=1)
```

```python
ms10avg = np.average(avg)
ms10err = np.std(avg)
```

## Plot

```python
_, ax = plt.subplots(figsize=(10, 10))

ax.errorbar(x=[1, 10, 100], y=[s1avg, ms100avg, ms10avg], fmt='r--', yerr=[s1err, ms100err, ms10err])

ax.set_title('Latency')
ax.set(ylabel='Average Delay (s)', xlabel='Packet Rate (pps)')
```
